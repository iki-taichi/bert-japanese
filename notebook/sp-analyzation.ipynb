```python
with open('wiki-ja.vocab', 'r') as f:
    texts = [_.strip() for _ in f.readlines()]
token_score_pairs = [_.split('\t') for _ in texts]
```

```python
# 登録トークン数
len(token_score_pairs)
```

```python
tokens = set([p[0] for p in token_score_pairs])
```

```python
# トークンの表層でソートして列挙
for t, _ in sorted(token_score_pairs, key=lambda p:p[0]):
    print(t)
```

```python
# 文字列の後ろからソートすると?
for t, _ in sorted(token_score_pairs, key=lambda p:p[0][::-1]):
    print(t)
```

```python
#
# [文字列をそのままソートした場合]
# - ▁で始まっているトークンがある
# - 、で始まっているトークンがある
# - 助詞助動詞・定型表現が前についているパターンがある
#   が~、に~、の~、を~、と~、とは~、こと~、として~、は~、も~、から~、で~、する~、し~、され~、という~、て~、まで~、へ~、(る~、ら~)
#
# [反転させてソートした場合]
# - 1/3程度は活用の途中だったり、 助詞助動詞・定型表現が後ろについているパターン
#  ~は、~が、~も、~から、など
# - 前からソートしたのと合わせると、前後への機能語の付きやすさにはっきりとした偏りはなさそう
# - カタカナのトークンには機能語があまりくっついていない
# - 文節に分ける感覚とも少し違う
#
```

```python
head_patterns = ['が', 'に', 'の', 'を', 'と', 'とは', 'こと', 'として', 'は', 'も', 'から', 'で', 'する', 'し', 'され', 'という', 'て', 'まで', 'へ', 'る', 'ら']
print('%s\t%s\t%s'%('表層', '始', '終'))
for h in head_patterns:
    c_head = 0
    c_tail = 0
    for t in tokens:
        if t.startswith(h):
            c_head += 1
        if t.endswith(h):
            c_tail += 1
    print('%s\t%d\t%d'%(h, c_head, c_tail))
```

```python
import re        
```

```python
#アルファベットだけでできているトークン
only_alphabets = list(filter(lambda p: re.search('[^a-z]', p[0].lower()) is None, token_score_pairs))
for t, _ in sorted(only_alphabets):
    print(t)
```

```python
# アルファベットだけでできているトークンもそこそこ含まれている
print('len_only_alphabets', len(only_alphabets))
```

```python
# ▁を含むトークン
contain_u2581 = list(filter(lambda p: p[0].count('\u2581')>0, token_score_pairs))
for t, _ in sorted(contain_u2581):
    print(t)
```

```python
print('len_contain_u2581', len(contain_u2581))
```

```python
# すべての▁を含むトークンは先頭に▁を持っているか？
print(all([p[0].startswith('\u2581') for p in contain_u2581]))
```

```python
# 先頭の▁を取り除いたとき、残りの部分はtokensに含まれるか？
not_contained = []
for t, _ in sorted(contain_u2581):
    target = t.strip('\u2581')
    if not (target in tokens):
        not_contained.append(target)
        print(target)
```

```python
print(len(not_contained))
```

```python
# 、を含むトークン
contain_u3001 = list(filter(lambda p: p[0].count('\u3001')>0, token_score_pairs))
for t, _ in sorted(contain_u3001):
    print(t)
```

```python
print('len_contain_u3001', len(contain_u3001))
```

```python
# 、を取り除いたとき、残りの部分はtokensに含まれるか？
not_contained_u3001 = []
for t, _ in sorted(contain_u3001):
    target = t.replace('\u3001', '')
    if not (target in tokens):
        not_contained_u3001.append(target)
        print(target)
```

```python
# * 先頭が▁のトークン
#     * ▁を取り除いてもトークンが存在するものはそのまま
#     * ▁を取り除くとトークンがなくなるものは、辞書のエントリーを▁を取り除いたものに置換
# * 先頭が、のトークン
#     * ほぼ辞書に含まれるのでそのまま
# * 助詞助動詞を含むトークン
#     * 想像以上にバリエーションがあるのでいったんそのまま
```
